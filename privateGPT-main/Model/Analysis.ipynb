```python
sns.boxplot(data, x= "Rating", orient="h")
plt.title("Box Plot of Rating")
```

```python
# Identify outliers count : 
# Calculate the IQR
Q1 = data['Rating'].quantile(0.25)
Q3 = data['Rating'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("---- Rating ----")
print("  Lower Bound :", max(lower_bound, 0))
print("  Upper Bound :", min(upper_bound, 10))
# % outliers from the dataset
print("  Percentage of outliers in the dataset : ", (1 - len(data[(data['Rating'] >= lower_bound) & (data['Rating'] <= upper_bound)])/len(data))*100, "%")
print("> Considering a Normalisation Method for the Rating")
```
